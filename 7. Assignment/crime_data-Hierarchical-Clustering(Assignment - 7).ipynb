```python
# Importing Libraries
import pandas as pd    
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading Dataset
Crime = pd.read_csv("crime_data.csv")   
Crime
```

```python
Crime.rename({"Unnamed: 0" : "Region"}, axis = 1 ,inplace = True)
Crime
```

```python
Crime.info()
```

<font color = 'teal'>There are no null values, so we can move furether with the dataset</font>

```python
#Normalise the data
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
Crime_1 = norm_func(Crime.iloc[:,1:])
Crime_1
```

```python
# create dendrogram
plt.figure(figsize=(10,8))
Dendogrm=sch.dendrogram(sch.linkage(Crime_1,method="average"))  # Create Dendogram
```

```python
# Create Clusters define Agglomerative
hc = AgglomerativeClustering(n_clusters=4, affinity="euclidean", linkage="average")
```

```python
y_hc=hc.fit_predict(Crime_1)         # Fit the data in the model use fit_predict
Clusters=pd.DataFrame(y_hc,columns=["clusters"])
```

```python
Clusters
```

```python
Crime["clusterid"] = Clusters
```

```python
Crime.groupby("clusterid").agg(["mean"]).reset_index()
```

```python
Crime[Crime["clusterid"] == 0] # Here we can see 0 cluster has lowest crime rate compared to others
```

```python
plt.figure(figsize = (10,7))
plt.scatter(Crime["clusterid"], Crime["Murder"], c = hc.labels_)
```
