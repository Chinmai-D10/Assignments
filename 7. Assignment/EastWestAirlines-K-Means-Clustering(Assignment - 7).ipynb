```python
# Imorting Libraries
import pandas as pd   
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading Data
Airline = pd.read_excel("EastWestAirlines.xlsx", sheet_name = 1 )  
Airline 
```

```python
Airline_1 = Airline.drop(['ID#'], axis =1)
Airline_1
```

```python
Airline_1.info()
```

```python
# Normalisation Function 
Airline_1_norm = StandardScaler().fit_transform(Airline_1)
```

```python
# How to find optimum number of  cluster
#The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion:
```

```python
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(Airline_1_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,11),wcss,"o--g")
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# Build Cluster algorithm using K=4

```python
# Cluster algorithm using K=4
clusters4=KMeans(4,random_state=30).fit(Airline_1_norm)
clusters4
```

```python
clusters4.labels_      # This will show clusters created
```

```python
Airline_2=Airline_1.copy()    # Assign clusters to the data set
Airline_2['clusters4id']=clusters4.labels_
Airline_2
```

```python
clusters4.cluster_centers_   # Compute the centroids for K=4 clusters with 11 variables
```

```python
Airline_2.groupby('clusters4id').agg(['mean']).reset_index()  # Group data by Clusters (K=4)
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(Airline_2['clusters4id'],Airline_2['Balance'], c = clusters4.labels_) 
```
